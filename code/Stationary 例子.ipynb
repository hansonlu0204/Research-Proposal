```python
import numpy as np
import pandas as pd
import statsmodels
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import coint
plt.style.use('seaborn')
```

```python
np.random.seed(100)
x = np.random.normal(0, 1, 500)
y = np.random.normal(0, 1, 500)
X = pd.Series(np.cumsum(x)) + 100
Y = X + y + 30
for i in range(500):
    X[i] = X[i] - i/10
    Y[i] = Y[i] - i/10
    
X.plot()
Y.plot()
plt.xlabel("Time")
plt.ylabel("Price")
plt.legend(["X", "Y"])
```

```python
(Y-X).plot()
plt.axhline((Y-X).mean(), color="red", linestyle="--")
plt.xlabel("Time")
plt.ylabel("Price")
plt.legend(["Y-X", "Mean"])
```
